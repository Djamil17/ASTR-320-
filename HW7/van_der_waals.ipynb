```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy.optimize import curve_fit
import scipy.stats as st    #for chi-squared probability

def chi_squared(theory:np.array,data:np.array,sigma:np.array)->np.array:
    """
    This function calculates TOTAL chi-squared between Theory and Data using sigma 
    as errors.
    The 3 arrays must be of equal size.
    Note: This is NOT reduced chi-squared
    Usage:  
     inputs: theory = input hypothesis (or Theory)
             data = Data points
             sigma = uncertainty on data points
     output: if arrays are of equal size returns the TOTAL chi-squared
             if arrays are not of equal size returns -1.0
    """
    if np.size(theory)==np.size(data) and np.size(data)==np.size(sigma):
        return np.sum((theory-data)**2/sigma**2)
    else:
        print('error - arrays of unequal size')
        return -1.


T=list(range(100,700,100))
B=[-160,-35,-4.2,9,16.9,21.3]
B_T=lambda x,a,b:b-a/x  

p0 = 1.0,1.0
poptl, pcovl=curve_fit(B_T,T,B,p0)
a,b=poptl[0],poptl[1]
B_predicted=[B_T(t,a,b) for t in range(100,700,10)]

plt.figure(figsize=(12,5))
plt.title(r"Van Der Waals Virial Cofficient B(T): Best-Fit using $\chi^2$ Minimization")
plt.xlabel("Temperature (K)")
plt.ylabel(r"B ($\frac{cm^3}{mol}$)")
plt.scatter(T,B,label='Data',color='orange')
plt.plot(range(100,700,10),B_predicted,label=r'Best-Fit: B(T)=%5.2f - $\frac{%5.2f}{RT}$' % (b,a))
plt.legend()
plt.show()

B=np.array(B)
B_predicted=np.array(B_T(T,a,b))
error=np.sqrt((B-B_predicted)**2)
chi2=chi_squared(B_T(T,a,b),B,error)
dof = np.size(T)-2#degrees of freedom = data points - #parameters (here 2)
prob = st.chi2.sf(chi2,dof) #chi-squared probability from scipy function

print(f"chi-squared is: {chi2}")
print(f"dof is: {dof}")
print(f"χ-mu is: {chi2/dof}")
print(f"prob is: {prob}")
```

The Van der Waals equation *seems* to fit the data quite well. I did an analysis of the χ-mu and found that it was close to 1.5, giving a quantitative measure of closeness, where 1 is interpreted often as meaning extremely close. That being said the probability was low .2, only 20% of the time do errors owe purely to chance as opposed to the error introduced by the model. This probably has to do with the crude residual I used for error. Whatever the case I think it is fair to say that our model reproduces the data quite well, at least in shape. The Van der Waals equation can be seen as a fairly accurate description in this regime.
