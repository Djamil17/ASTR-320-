```python
from math import log10
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
## build data 
au=1.496e11
years=3.154e7
planet_labels=['Mercury','Venus','Earth','Mars','Jupiter','Saturn','Uranus','Neptune']
semi_major_axis=[.3870993,.723336,1.000003,1.52371,5.2029,9.537,19.189,30.0699]
semi_major_axis=[au*x for x in semi_major_axis]
period=[0.2408467,0.61519726,1.0000174,1.8808158,11.862615,29.447498,84.016846,164.79132]
period=[years*x for x in period]
semi_major_axis=[log10(x) for x in semi_major_axis]
period=[log10(x) for x in period]
period_arr,semi_major_axis_arr=np.array(period),np.array(semi_major_axis)
## cite : all data from https://www.princeton.edu/~willman/planetary_systems/Sol/ data was collected from cite in terms of au
## and in terms of years and converted to seconds and meters

```

```python
## ploting now 
a, k = np.polyfit(semi_major_axis_arr,period_arr, 1)
plt.scatter(semi_major_axis_arr, period_arr)
plt.plot(semi_major_axis_arr, a*semi_major_axis_arr+k,label=f'y={round(a,3)}x+{round(k,3)}',color='red')
plt.title('Keplers Law: Log10(Period) vrs. Log10(Semi-Major Axis)')
plt.xlabel('Semi-Major Axis Log10(meters)')
plt.ylabel('Period Log10(seconds)')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
