```python
"""


"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

Β=[.01,.1,.2]
h=6.62607015e-34 

fermi_dirac=lambda ξ,β: 2/h**3 *1/((np.exp((ξ-1)/β))+1)

x=np.linspace(0,5,1000)
plt.figure(figsize=(12,6))
for β in Β: 
    plt.plot(x,fermi_dirac(x,β),label=r'$\beta$='+f"{β}")

plt.xlabel(r'$\xi$ $(\frac{E}{E_f})$')
plt.ylabel(r'$\mathfrak{F}(\xi)$')
plt.title(r'$\mathfrak{F}(\xi)=\frac{2}{h^3}\frac{1}{e^((\xi-1)/\beta)}$ where $\xi=\frac{E}{E_f}$,$\beta=\frac{kT}{E_f}$')
plt.legend()
plt.grid('--')
plt.show()
```

```python
for β in Β:
    β_0=β
    I1=lambda ξ: (3/2)*((ξ**(1/2))/(np.exp((ξ-1)/β_0)+1))
    I3=lambda ξ: (5/2)*((ξ**(3/2))/(np.exp((ξ-1)/β_0)+1))
    I3_result=quad(I3,0,np.Infinity)[0] #integrates from 0 to infinity
    I1_result=quad(I1,0,np.Infinity)[0]**(5/3)
    ratio=I3_result/I1_result
    print(r"β:",β)
    print(r"ratio:",ratio)
    print()
```
